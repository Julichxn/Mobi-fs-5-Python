**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
thres = 0.1
H = img >= thres
plt.figure()
plt.imshow(H, cmap = "grey")
plt.colorbar()
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
G = plt.imread('data/NIH3T3/gt/0.png')
plt.imshow(G)
type(G)
```

Compute and print the Dice coefficient:

```python
IGI = np.sum(G)
GnH = np.sum(G * H)
IHI = np.sum(H)
Nenner = IHI +IGI

Z = (2*(GnH))/Nenner
print(Z)
```
